# 4.

```python
car_sales=pd.read_csv("car-sales.csv")
car_sales.head()
```

```python
plt.plot(car_sales['Price'])
plt.show()
```

```python
plt.plot(car_sales['Odometer (KM)'])
plt.show()
```

```python
# They Both are able to be graphed with the plot() function becasue they are numerical 
```

# 5.

```python
car_sales[car_sales.columns[4]] = car_sales[car_sales.columns[4]].replace('[\$,]', '', regex=True).astype(int) 
#use int instead of int64 because they are =
```

```python
car_sales.head()
```

# 6.

```python
car_sales['Total Sales']=car_sales['Price'].cumsum()
```

```python
car_sales.head()
```

# 7.

```python
car_sales['Sale Date']=pd.date_range(start ='2-21-2022', periods=len(car_sales))
```

```python
car_sales.head()
```

# 8.

```python
plt.plot(car_sales['Price'],car_sales['Total Sales'])
plt.title("Price vs. Total Sales Line Plot")
plt.xlabel('Price')
plt.ylabel('Total Sales')
plt.show()
```

# 9.

```python
plt.scatter(car_sales['Odometer (KM)'], car_sales['Price'])
plt.title('Odometer vs. Price Scatter Plot')
plt.xlabel('Odometer (KM)')
plt.ylabel('Price')
plt.show()
```

# 10.

```python
plt.bar(car_sales['Make'], car_sales['Odometer (KM)'])
plt.title("Make vs. Odometer Bar Plot")
plt.xlabel("Make")
plt.ylabel('Odometer (KM)')
plt.show()
```

# 11.

```python
plt.hist(car_sales['Make'])
plt.title('Histogram of Car Make')
plt.xlabel('Make')
plt.show()
```

# 12.

```python
fig, ((ax1, ax2), (ax3, ax4))=plt.subplots(2,2, figsize=(12,10))
# Line for Price vs Total sales
ax1.plot(car_sales['Price'],car_sales['Total Sales'])
ax1.set_title("Price vs. Total Sales Line Plot")
ax1.set_xlabel('Price')
ax1.set_ylabel('Total Sales')

# Scatter for Odometer vs Price
ax2.scatter(car_sales['Odometer (KM)'], car_sales['Price'])
ax2.set_title('Odometer vs. Price Scatter Plot')
ax2.set_xlabel('Odometer (KM)')
ax2.set_ylabel('Price')

# Bar graph for Make vs Odometer
ax3.bar(car_sales['Make'], car_sales['Odometer (KM)'])
ax3.set_title("Make vs. Odometer Bar Plot")
ax3.set_xlabel("Make")
ax3.set_ylabel('Odometer (KM)')

# Histogram of car Make
ax4.hist(car_sales['Make'])
ax4.set_title('Histogram of Car Make')
ax4.set_xlabel('Make')

plt.show()
```

# 16.

```python
heatmap=car_sales[['Total Sales','Price','Odometer (KM)']].corr(method='pearson')
cols=['Total Sales','Price','Odometer (KM)']
axis=sns.heatmap(heatmap, annot=True, xticklabels=cols, yticklabels=cols)
```

```python
# There is not a very strong relationship between any of the columns. 
# The strongest relationship is between the Price and Total sales at .16
```
